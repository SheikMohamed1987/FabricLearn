```python
df.write.format("delta").saveAsTable("external_products", path="abfss://8a7517bb-4187-4d42-a0a5-3886bf0786d6@onelake.dfs.fabric.microsoft.com/0bda76b3-c969-41da-a95c-fa5bbb5fb164/Files/external_products")
```

```python
%%sql

CREATE TABLE products
USING DELTA
LOCATION 'Files/external_products';
```

```python
%%sql

UPDATE products
SET ListPrice = ListPrice * 0.9
WHERE Category = 'Mountain Bikes';
```

```python
%%sql

DESCRIBE HISTORY products;
```
